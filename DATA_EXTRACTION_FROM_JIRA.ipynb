```python
# THIS SCRIPT WILL CHANGE JIRA HTML DATA TO EXCEL WORKSHEET BASED ON CERTAIN EXCEL TEMPLATE

import xlwings as xw
import re
from bs4 import BeautifulSoup

def readRAWHTML():
    # OPEN LOCAL HTML FILE, RAW HTML DATA
    HTMLFileToBeOpened = open(r"C:\Users\h638206\Downloads\ALL OSIRIS NC TO SG (JIRA ENGINEERING) 2022-06-24T04_47_48+0200.html",'r', encoding='utf-8')

    # READING THE FILE AND STORING IN A VARABLE
    contents = HTMLFileToBeOpened.read()

    soup = BeautifulSoup(contents, 'lxml')

    # FIND SUMMARY
    summarys = soup.find_all('td', class_='summary')
    summary = []
    for x in summarys:
        summary.append((re.findall(r'\n[ ]+(.*)', x.get_text())))

    # FIND KEY PARENT
    key_parents = soup.find_all('td', class_='customfield_12617')
    key_parent = []
    for x in key_parents:
        key_parent.append(re.findall(r'[ ]+(.*)', x.get_text()))

    # FIND ROOT CAUSE
    root_causes = soup.find_all('td', class_='customfield_10504')
    root_cause = []
    for x in root_causes:
        root_cause.append(re.findall(r'[ ]+(.*)', x.get_text()))

    # FIND DESCRIPTION
    descriptions = soup.find_all('td', class_='description')
    description = []
    # for x in descriptions:
    #     description.append(re.findall(r'[ ]+(.*)', x.get_text()))
    for x in descriptions:
          description.append(x.get_text())

    # FIND AFFAIR
    affairs = soup.find_all('td', class_='customfield_13102')
    affair = []
    for x in affairs:
        affair.append(re.findall(r'(.*)\n', x.get_text()))

    # FIND CATALOG EQUIPMENT
    catalog_equipments = soup.find_all('td', class_='customfield_10013')
    catalog_equipment = []
    for x in catalog_equipments:
        catalog_equipment.append(re.findall(r'(.*)\n', x.get_text()))

    # FIND EQUIPMENT
    equipments = soup.find_all('td', class_='customfield_10506')
    equipment = []
    for x in catalog_equipments:
        equipment.append(re.findall(r'(.*)\n', x.get_text()))
    
    # FIND STATUS
    statuss = soup.find_all('td', class_='status')
    status = []
    for x in statuss:
        status.append(re.findall(r'\n(.*)', x.get_text()))
        
    # FIND EQUIPMENT COMPONENT 
    components = soup.find_all('td', class_='customfield_11400')
    component = []
    for x in components:
        component.append(x.get_text())

    # USE RE METHOD
    # ALL PATTERNS THAT TO MATCH
    pattern_key = r'data-issue-key="(ANO-\d+)"'
    pattern_created = r'<td class="created"> (\d+/\w+/\d+)'
    pattern_issuetype = r'<td class="issuetype">[ ]+(.*)'
    pattern_assignee = r'<td class="assignee">[ ]+(.*)'
    pattern_reporter = r'<td class="reporter">[ ]+(.*)'


    key = re.findall(pattern_key, contents)
    issuetype = re.findall(pattern_issuetype, contents)
    created = re.findall(pattern_created, contents)
    issuetype = re.findall(pattern_issuetype, contents)
    assignee = re.findall(pattern_assignee, contents)
    reporter = re.findall(pattern_reporter, contents)

    print('key ', len(key))
    print('key_parent ', len(key_parent))
    print('issuetype ', len(issuetype))
    print('root_cause ', len(root_cause))
    print('summary ', len(summary))
    print('description ', len(description))
    print('status ', len(status))
    print('assignee ', len(assignee))
    print('reporter ', len(reporter))
    print('created ', len(created))
    print('affair ', len(affair))
    print('catalog_equipment ', len(catalog_equipment))
    print('equipment ', len(equipment))
    print('component ', len(component))

    return key, key_parent, issuetype, root_cause, summary, description, status, assignee, reporter, created, affair, catalog_equipment, equipment, component


# WRITE ONE LINE PASS TWO PARAMETERS: LINE NUMBER: j, ITEM NUMBER: i
def writeOneLine(item_number, line_number, key, key_parent, issuetype, root_cause, summary, description, 
                 status, assignee, reporter, created, affair, catalog_equipment, equipment, component, wb):
    
    j=line_number
    i=item_number
    ws = wb.sheets['NC']

    table_headers={'KEY': 'A', 'KEY(PARENT)': 'B', 'ISSUE TYPE': 'C', 'ROOT CAUSE': 'D', 'SUMMARY': 'E', 'DESCRIPTION': 'F', 
               'STATUS': 'G', 'ASSIGNEE': 'H', 'REPORTER': 'I', 'CREATED': 'J', 'AFFAIR': 'K', 'CATALOG EQUIPMENT': 'L', 
               'EQUIPMENT': 'M', 'Equipment Component': 'N'}
    
    for table_header, col in table_headers.items():
        if table_header == 'KEY':
            ws.range(col+str(j)).value = key[i]
        if table_header == 'KEY(PARENT)':
            ws.range(col+str(j)).value = key_parent[i]
        if table_header == 'ISSUE TYPE':
            ws.range(col+str(j)).value = issuetype[i]
        if table_header == 'ROOT CAUSE':
            ws.range(col+str(j)).value = root_cause[i]
        if table_header == 'SUMMARY':
            ws.range(col+str(j)).value = summary[i]
        if table_header == 'DESCRIPTION':
            ws.range(col+str(j)).value = description[i]
        if table_header == 'STATUS':
            ws.range(col+str(j)).value = status[i]
        if table_header == 'ASSIGNEE':
            ws.range(col+str(j)).value = assignee[i]
        if table_header == 'REPORTER':
            ws.range(col+str(j)).value = reporter[i]
        if table_header == 'CREATED':
            ws.range(col+str(j)).value = created[i]
        if table_header == 'AFFAIR':
            ws.range(col+str(j)).value = affair[i]
        if table_header == 'CATALOG EQUIPMENT':
            ws.range(col+str(j)).value = catalog_equipment[i]
        if table_header == 'EQUIPMENT':
            ws.range(col+str(j)).value = equipment[i]
        if table_header == 'Equipment Component':
            ws.range(col+str(j)).value = component[i]

def main():
    # READ RAW DATA
    key, key_parent, issuetype, root_cause, summary, description, status, assignee, reporter, created, affair, catalog_equipment, equipment, component = readRAWHTML()
    
    # OPEN EXCEL, THE FILE THAT YOU WANT TO MODIFY
    wb = xw.Book(r"C:\Users\h638206\Downloads\OSIRIS_JIRA_NC_FOLLOW_UP20220624.xlsx")
    ws = wb.sheets['NC']
    
    # READ ALL ANO NUMBER
    addr = ws.range('A1').expand('down').address
    anolist = ws.range(addr).value
    
    # CALL FUNCTION AND WIRTE IN EXCEL
    for i in range(len(key)):
        if key[i] in anolist:
            line_number = anolist.index(key[i])+1
            item_number = i
            writeOneLine(item_number, line_number, key, key_parent, issuetype, root_cause, summary, description, 
                 status, assignee, reporter, created, affair, catalog_equipment, equipment, component, wb)
        else:
            # RETURN LAST EMPTY LINE
            line_number = int((re.findall(r'(\d+)\Z', ws.range('A1').expand('down').address))[0]) + 1
            item_number = i
            writeOneLine(item_number, line_number, key, key_parent, issuetype, root_cause, summary, description, 
                 status, assignee, reporter, created, affair, catalog_equipment, equipment, component, wb)
        print(f'{key[i]} has been listed in Excel. Total counts {i+1}')

if __name__ == '__main__':
    main()
```
